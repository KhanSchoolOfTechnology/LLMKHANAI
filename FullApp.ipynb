```python
import openai
from kivy.uix.screenmanager import Screen
from kivy.uix.scrollview import ScrollView
from kivy.uix.boxlayout import BoxLayout
from kivymd.uix.list import MDList, OneLineListItem
from kivymd.uix.textfield import MDTextField
from kivymd.uix.button import MDRaisedButton
from kivymd.app import MDApp
from kivymd.uix.toolbar import MDTopAppBar
from kivy.utils import get_color_from_hex


class ChatScreen(Screen):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)

        # Create a ScrollView
        scroll_view = ScrollView()
        scroll_view.pos_hint = {'top': 0.9}
        scroll_view.size_hint = (1, 0.80)

        # Create an MDList for the conversation
        conversation_list = MDList()
        conversation_list.orientation = "lr-tb"

        # Create an MDTextField for user input
        layout = BoxLayout()
        layout.orientation = 'horizontal'
        layout.padding = 10
        layout.spacing = 10

        text_input = MDTextField(size_hint_x=0.6)
        text_input.hint_text = "Enter Text Here"
        text_input.width = 400

        # Create a button to send messages (You can customize this)
        send_button = MDRaisedButton(text="Send")
        send_button.size_hint_x = 0.4
        send_button.bind(on_release=self.send_message1)

        # Create an app bar
        app_bar = MDTopAppBar()
        app_bar.title = "Translation App"
        app_bar.pos_hint = {'top': 1}

        # Add widgets to the screen
        self.add_widget(app_bar)

        self.conversation_list = conversation_list
        scroll_view.add_widget(self.conversation_list)
        self.text_input = text_input
        self.add_widget(scroll_view)
        layout.add_widget(text_input)
        layout.add_widget(send_button)
        self.add_widget(layout)

    def send_message1(self, *args):
        # Get the user's message from the text input
        user_message = self.text_input.text
        self.text_input.text = ""  # Clear the input field

        # Create an MDLabel for the user's message and add it to the conversation list
        # message_label = OneLineListItem(text=user_message)
        # self.conversation_list.add_widget(message_label)

        # Get ChatGPT response
        chatgpt_response = self.get_chatgpt_response(user_message)

        label = MDTextField()
        label.text = chatgpt_response
        label.text_color = (0, 0, 1, 1)

        # Create an MDLabel for the ChatGPT response and add it to the conversation list
        response_label = OneLineListItem(text=chatgpt_response)
        label.multiline = True


        self.conversation_list.add_widget(label)

    def get_chatgpt_response(self, user_message):
        # Initialize OpenAI API with your API key
        openai.api_key = apikey

        # Send a request to ChatGPT for translation from English to Urdu
        response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo",
            messages=[
                {"role": "system", "content": f"Translate this english text {user_message} to italian"},
                {"role": "user", "content": user_message},
            ],
        )

        # Extract and return the assistant's reply
        chatgpt_response = response['choices'][0]['message']['content']
        return chatgpt_response



class ConversationApp(MDApp):
    def build(self):
        return ChatScreen()


if __name__ == "__main__":
    ConversationApp().run()
```
